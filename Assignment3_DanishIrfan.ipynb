```python
# Assignment 3

# Q1. Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

a = [10, -20, 30, -40, 50, -60, 0]
positive_count = 0
negative_count = 0
for num in a:
    if num >= 0:
        positive_count += 1
    elif num < 0:
        negative_count += 1
print("Positive count:", positive_count)
print("Negative count:", negative_count)
```

```python
# Q2. Using List Comprehension. Write a Python program to count positive and negative numbers in a list.
# a = [10, -20, 30, -40, 50, -60, 0]

a = [10, -20, 30, -40, 50, -60, 0]

positive = [num for num in a if num >= 0]
negative = [num for num in a if num < 0]

print("Positive Numbers :", len(positive))
print("Negative Numbers :", len(negative))
```

```python
# Q3. Remove Multiple Elements from List in Python
# a = [10, 20, 30, 40, 50, 60, 70]
# remove = [20, 40, 60]
# Resulting list = [10, 30, 50, 70]

a = [10, 20, 30, 40, 50, 60, 70]
remove = [20, 40, 60]

result = []

for i in a:
    if i not in remove:
        result.append(i)

print("Resulting list:", result)
```

```python
# Q4. Remove duplicates from list in Python
# a = [1, 2, 2, 3, 4, 4, 5]

a = [1, 2, 2, 3, 4, 4, 5]
result = []

result = list(set(a))

print("Resulting list:", result)
```

```python
# Q5. Python program to find the sum of all items in a dictionary
# Input: {'a': 100, 'b': 200, 'c': 300}
# Output: 600

input = {'a': 100, 'b': 200, 'c': 300}
sum = 0

for value in input.values():
    sum += value

print("Sum of all items in the dictionary:", sum)
```

```python
# Q6. Key with Maximum Unique Values - Python
# dict1 = {"key1": [5, 7, 5, 4, 5], "key2": [6, 7, 4, 3, 3], "key3": [9, 9, 6, 5, 5]}
# output : key2

dict1 = {"key1": [5, 7, 5, 4, 5], "key2": [6, 7, 4, 3, 3], "key3": [9, 9, 6, 5, 5]}

maxunique = 0
maxkey = ""

for key in dict1:
    unique = len(set(dict1[key]))
    if unique > maxunique:
        maxunique = unique
        maxkey = key

print("Key with maximum unique values:", maxkey)
```

```python
# Q7. Story: An online store records daily product sales.
# Convert this list into total sales per product.
# Output : {'apple': 8, 'banana': 3, 'orange': 4}

sales = [
    ("apple", 3),
    ("banana", 2),
    ("apple", 5),
    ("banana", 1),
    ("orange", 4)
]

total_sales = {}

for product, quantity in sales:
    if product in total_sales:
        total_sales[product] += quantity
    else:
        total_sales[product] = quantity

print(total_sales)
```

```python
# Q8. Undo Feature in Text Editor
# A text editor supports an undo operation.
# Given operations:
# actions = ["type A", "type B", "delete", "type C"]
# Output : ['type A', 'type C']

# actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]
# Output : ['type A', 'type C', 'type A']

actions = ["type A", "type B", "delete", "type C"]
new_actions = []

for action in actions:
  if action == "delete":
    new_actions.pop()
  else:
    new_actions.append(action)

print(new_actions)

actions = ["type A", "type B", "delete", "type C","type B","delete","type A","type C","delete"]
new_actions = []
for action in actions:
  if action == "delete":
    new_actions.pop()
  else:
    new_actions.append(action)

print(new_actions)
```

```python
# Q9. Print the index number of those Two boxes whose weights sum to a target.
# weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
# target = 9

# Output
'''
0 1
4 5
0 6
4 7
0 8
8 9
'''

weights = [2, 7, 11, 15 , 6 , 3, 7 , 3 , 7 , 2]
target = 9

result = {}
for i, x in enumerate(weights):
  if (target - x) in result:
    print(result[target - x], i)
  result[x] = i
```

```python
# Q10. Online Store Price Update (map + lambda)
# An online store increases all product prices by 10%.
# prices = [100, 200, 300, 400]
# output [110, 220, 330, 440]

prices = [100, 200, 300, 400]
new_prices = list(map(lambda x: int(x * 1.1), prices))
# lambda is a short function without name
# lambda parameter : expression
print(new_prices)
```

```python
# Q11. Employee Bonus System (map + if–else)
# Employees earning ≥ 50,000 get a 20% bonus, others get 10%.
# salaries = [40000, 60000, 50000, 30000]

salaries = [40000, 60000, 50000, 30000]
# map() is to modify data
result = list(map(lambda x: int(x * 1.2) if x >= 50000 else int(x * 1.1), salaries))
print(result)
```

```python
# Q12. Order Filtering System (filter + multiple conditions)
# An e-commerce system wants orders above ₹1000 and not cancelled.

orders = [
    {"amount": 1200, "status": "completed"},
    {"amount": 800, "status": "completed"},
    {"amount": 1500, "status": "cancelled"},
    {"amount": 2000, "status": "completed"}
]
# filter is to filter data
valid_orders = list(filter(lambda x: x["amount"] > 1000 and x["status"] == "completed", orders))
print(valid_orders)
```

```python
# Q13. Data Cleaning Pipeline (map + filter + lambda)
# From a data pipeline:
#  a.)Remove negative numbers
#  b.) Square remaining numbers
#  c.) Keep only results > 50

# data = [-4, -1, 2, 4, 8]
# Output : [64]

data = [-4, -1, 2, 4, 8]
result = list(filter(lambda x: x > 50,filter(lambda x: x>0, map(lambda x: x**2, data))))
print(result)
```
